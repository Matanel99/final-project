```python
import os, re, pandas as pd
from IPython.display import Image, display

# 1) ××¦×‘×™×¢×™× ×œ×ª×™×§×™×™×ª ×”-val ×”×§×™×™××ª
exp_dir = "/content/drive/MyDrive/final_dataset_for_signs_model/update database 08.2025/yolov5/runs/val/exp3"
assert os.path.isdir(exp_dir), f"×œ× × ××¦××” ×”×ª×™×§×™×™×”: {exp_dir}"
print("Using:", exp_dir)

# 2) ×§×•×¨××™× ××ª ×§×•×‘×¥ ×”×œ×•×’ ×©×œ YOLOv5 ×•××—×œ×¦×™× ××ª ×˜×‘×œ×ª per-class
results_txt = os.path.join(exp_dir, "results.txt")
if not os.path.isfile(results_txt):
    raise FileNotFoundError("×œ× × ××¦× results.txt ×‘×ª×™×§×™×™×ª ×”-val. ×•×“×/×™ ×©×”×¨×¦×ª val.py ×¢× --verbose/--plots")

with open(results_txt, "r", encoding="utf-8", errors="ignore") as f:
    txt = f.read()

# YOLOv5 ××“×¤×™×¡ ×˜×‘×œ×” ×¢× ×›×•×ª×¨×ª: Class Images Instances P R mAP50 mAP50-95
pat = r"(?:Class\s+Images\s+Instances\s+P\s+R\s+mAP50\s+mAP50-95\n)([\s\S]+?)(?:\n\n|$)"
m = re.search(pat, txt)
if not m:
    raise ValueError("×œ× × ××¦××” ×˜×‘×œ×ª per-class ×‘×ª×•×š results.txt (×™×ª×›×Ÿ ×©×”×œ×•×’ ×©×•× ×”).")

table_block = m.group(1).strip()
rows = [r for r in table_block.splitlines() if r.strip()]

data = []
for r in rows:
    toks = r.split()
    # ×‘×¡×•×£ ×›×œ ×©×•×¨×” ×™×© 6 ×¢×¨×›×™× ××¡×¤×¨×™×™×: Images, Instances, P, R, mAP50, mAP50-95
    nums = list(map(float, toks[-6:]))
    cls_name = " ".join(toks[:-6])  # ×©× ×”××—×œ×§×” (×™×›×•×œ ×œ×”×›×™×œ ×¨×•×•×—×™×)
    data.append([cls_name] + nums)

df = pd.DataFrame(data, columns=["class","images","instances","P","R","mAP@0.5","mAP@0.5:0.95"])
display(df)

# 3) ×©××™×¨×ª ×˜×‘×œ×ª ×”××˜×¨×™×§×•×ª ×œ-CSV ×‘××•×ª×” ×ª×™×§×™×™×”
out_csv = os.path.join(exp_dir, "per_class_metrics.csv")
df.to_csv(out_csv, index=False)
print("Saved CSV:", out_csv)

# 4) ××¦×™×’×™× ×’×¨×¤×™× ×©×™××•×©×™×™× ×× ×§×™×™××™×
for fn in ["confusion_matrix.png", "PR_curve.png", "F1_curve.png", "P_curve.png", "R_curve.png"]:
    path = os.path.join(exp_dir, fn)
    if os.path.isfile(path):
        print(fn)
        display(Image(filename=path))
```
