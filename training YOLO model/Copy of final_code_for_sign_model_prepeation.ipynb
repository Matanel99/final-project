```python
#extract the information from the xml paths
def extract_text(filename): #the function get the path to the file
  tree = et.parse(filename) #open the file that the function get
  root = tree.getroot()

  #extract filename:
  image_name = root.find('filename').text
  #width and hieght of the image
  width = root.find('size').find('width').text
  height = root.find('size').find('height').text
  parser = []
  #extract object name, xmin, xmax, ymin, ymax
  for obj in root.findall('object'):
    object_name = obj.find('name').text
    if object_name == '3':
      object_name = 'stop_sign'
    elif object_name == 'crosswalk-dx5v':
      object_name = 'crosswalk_sign'
    elif object_name == '9':
      object_name = 'crosswalk_sign'
    elif object_name == 'green':
      object_name = 'green_light'
    elif object_name == 'red':
      object_name = 'red_light'
    elif object_name == 'yellow':
      object_name = 'yellow_light'
    elif object_name == 'Traffic-Light':
      object_name = 'traffic_light'
    elif object_name == 'Stop-Sign':
      object_name = 'stop_sign'
    elif object_name == 'Crosswalk':
      object_name = 'crosswalk_sign'


    xmin = obj.find('bndbox').find('xmin').text
    xmax = obj.find('bndbox').find('xmax').text
    ymin = obj.find('bndbox').find('ymin').text
    ymax = obj.find('bndbox').find('ymax').text
    parser.append([image_name, width, height, object_name, xmin, xmax, ymin, ymax])
  return parser
```

```python
#send all the xml files to "extract_text" function
parser_all = list(map(extract_text, xml_list)) #the function map oparate the function 'extract_text' for each variable in xml list. in the end we get new list
for i in parser_all:
  print(i)
```

```python
data=reduce(lambda x,y : x+y, parser_all)
for i in data:
  print(i)
```

```python
df = pd.DataFrame(data, columns=['filename', 'width', 'height', 'object_name', 'xmin', 'xmax', 'ymin', 'ymax'])
df.head()
```

# convert the data to YOLO format

```python
'''
we need to convert the data to YOLO format:
<object-class> <x_center> <y_center> <width> <height>
when x_center and y_center are normalized by the image width and height.
width and height are normalized by the total width and height of the image.
'''
```

```python
df.info() # present info about df: number of columns and rows, manes of the columns and the type of them
```

```python
#we can see above that all sizes are in type 'object', we need to convert them to integer types
cols = ['width', 'height', 'xmin', 'xmax', 'ymin', 'ymax']
df[cols] = df[cols].astype('int')
df.info()
```

```python
#we need to convert the bndbox to x_center and y_center with the formula: x_center= ((xmin+xmax)/2)/width
df['x_center'] = ((df['xmin'] + df['xmax'])/2)/df['width']
df['y_center'] = ((df['ymin'] + df['ymax'])/2)/df['height']
#we also need to convert the width and hieght with the formula: width = (xmax-xmin)/width_of_the-image
df['w'] = (df['xmax'] - df['xmin'])/df['width']
df['h'] = (df['ymax'] - df['ymin'])/df['height']

df.head()
```

```python
#label encoding
def label_encoding(x):
  labels = {'red_light':0, 'green_light':1, 'yellow_light':2,'crosswalk_sign':3 , 'stop_sign': 4    }
  return labels[x]
```

```python

def save_data(filename, folder_path, group_obj, source_folders):
  # נסה למצוא את הקובץ בכל אחת מהתיקיות
  src = None
  for folder in source_folders:
      potential_src = os.path.join(folder, filename)
      if os.path.exists(potential_src):
          src = potential_src
          break

  if not src:
      print(f"File {filename} does not exist in any source folder. Skipping...")
      return  # Skip to the next file

  # העברת הקובץ ליעד
  dst = os.path.join(folder_path, filename)
  copy(src, dst)

  # שמירת התוויות
  text_file = os.path.join(folder_path, os.path.splitext(filename)[0] + '.txt')
  group_obj.get_group(filename).set_index('filename').to_csv(
      text_file, sep=' ', index=False, header=False
  )
```
